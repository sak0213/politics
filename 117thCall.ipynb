```python
import requests
import pandas as pd
from key_data import open_secret_key
import time
```

```python
CID_List = pd.read_csv("117thCongress.csv")
CIDs = CID_List["CID"].unique()
```

```python
url = "http://www.opensecrets.org/api/?method=getLegislators&"
# tester = requests.get(f'{url}id=pa&apikey={open_secret_key}&output=json').json()
id = []
last_name = []
first_name = []
bioguide = []
office = []
party = []
first_elected = []
birthdate = []
website = []
counter = 0
```

```python
tester = "N00003028"
request = requests.get(f'{url}id={tester}&apikey={open_secret_key}&output=json').json()
print(request['response']['legislator']['@attributes'])
```

```python
for cid in CIDs:
    try:
        req = requests.get(f'{url}id={cid}&apikey={open_secret_key}&output=json').json()
        xxx = req['response']['legislator']
        id.append(xxx['@attributes']['cid'])  
        last_name.append(xxx['@attributes']['lastname'])
        first_name.append(xxx['@attributes']['firstlast'])
        bioguide.append(xxx['@attributes']['bioguide_id'])
        office.append(xxx['@attributes']['office'])
        party.append(xxx['@attributes']['party'])
        first_elected.append(xxx['@attributes']['first_elected'])
        birthdate.append(xxx['@attributes']['birthdate'])
        website.append(xxx['@attributes']['website'])
        print(f'{cid}worked!')
        counter += 1
        print(counter)
        
    except:
        print(f'{cid}didnt work')
```

```python
df = pd.DataFrame({
    "CID":id,
    "Last Name":last_name,
    "First Last":first_name,
    "Bioguide ID":bioguide,
    "Office":office,
    "Party":party,
    "First Elected":first_elected,
    "Birth Date":birthdate,
    "Website": website
})
```
